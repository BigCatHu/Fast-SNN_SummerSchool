## MNIST

```python
import matplotlib.pyplot as plt

# 准备数据
x = ['SNNα', 'SNNβ', 'SNNγ']
# y1 = [78.92 ,98.90 ,98.89 ]
y1 = [96.98	,97.46 ,97.54 ]
# y2 = [14.62 ,62.72 ,81.78 ]
y2 = [97.10 ,98.11 ,98.18]
# threshold_value0 = 98.98
threshold_value0 = 97.75
# threshold_value1 = 98.93
threshold_value1 = 98.18
Architectures = "Vgg11"
i = 2

# 创建包含两个子图的图像对象，并设置网格规范
fig, axs = plt.subplots(1, 2, figsize=(6, 4), sharey=True,
                        gridspec_kw={'width_ratios': [0.5, 0.5]})

# 在第一个子图中创建柱状图，并添加数值标签
rects1 = axs[0].bar(x, y1, width=0.6, color='skyblue')
axs[0].axhline(threshold_value0, color='r', linestyle='--')
axs[0].set_title(Architectures)
axs[0].set_xlabel('(c)')
axs[0].set_ylabel('Accuracy (%)')

for rect in rects1:
    height = rect.get_height()
    axs[0].text(rect.get_x() + rect.get_width() / 2, height,
                f'{height}', ha='center', va='bottom')
    
axs[0].annotate(f'ANNs: {threshold_value0}', xy=(0, threshold_value0),
                xytext=(20, -40), textcoords='offset points',
                arrowprops=dict(arrowstyle='->', color='black'))

# 在第二个子图中创建柱状图，并添加数值标签
rects2 = axs[1].bar(x, y2, width=0.6, color='lightgreen')
axs[1].axhline(threshold_value1, color='r', linestyle='--')
axs[1].set_title(Architectures)
axs[1].set_xlabel('(d)')
axs[1].set_ylabel('Accuracy (%)')

for rect in rects2:
    height = rect.get_height()
    axs[1].text(rect.get_x() + rect.get_width() / 2, height,
                f'{height}', ha='center', va='bottom')
    
# 给整个图增加一个大标题
plt.suptitle('Performance comparison of different SNN methods on MNIST')

# 添加图例说明颜色和含义
legend_handles = [rects1[0], rects2[0]]
legend_labels = ['3-bit', '2-bit']

# 添加柱形图图例和虚线图例
legend_elements = legend_handles + [plt.Line2D([0], [0], color='red', linestyle='--', label='ANNs')]
axs[1].legend(handles=legend_elements, labels=legend_labels + ['ANNs'], 
              loc='upper left', bbox_to_anchor=(1.05, 1))

# 在第二个子图中标出虚线的值
axs[1].annotate(f'ANNs: {threshold_value1}', xy=(0, threshold_value1),
                xytext=(20, -40), textcoords='offset points',
                arrowprops=dict(arrowstyle='->', color='black'))

# 调整子图之间的间距
plt.tight_layout()

# 设置保存图像的分辨率
dpi = 300


# 保存图像为高分辨率图片
plt.savefig(f'chart_high_res_{i}.png', dpi=dpi)

# 显示图形
plt.show()
```
